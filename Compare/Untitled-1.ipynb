```python
import pickle
import matplotlib.pyplot as plt
```

```python
def plot_results(file_paths):
    for file_path in file_paths:
        # 读取数据
        with open(file_path, 'rb') as file:
            data = pickle.load(file)
        
        # 绘制目标函数值的变化
        plt.plot(data['fhist'], label=f"{data['algorithm_name']} (niter={data['niter']}, nfev={data['nfev']}, pdfo_nfev={data['pdfo_nfev']})")
    
    plt.xlabel('Iteration')
    plt.ylabel('Objective Value')
    plt.title('Objective Value vs. Iteration')
    plt.legend()
    plt.show()
```

```python
file_paths = [
    #'pkl/PDFO_1_result.pkl',
    #'pkl/PDFO_Single_1_result.pkl',
    'pkl/PDFO_2_result.pkl',
    'pkl/PDFO_Single_2_result.pkl', 
    #'pkl/PDFO_3_result.pkl',
    #'pkl/PDFO_Single_3_result.pkl',
    #'pkl/PDFO_4_result.pkl',
    #'pkl/PDFO_Single_4_result.pkl',
    'pkl/PDFO_2DS_1_result.pkl',
    'pkl/PDFO_Single_2DS_1_result.pkl',
    'pkl/PDFO_Single_10DS_1_result.pkl',
]
plot_results(file_paths)
```
